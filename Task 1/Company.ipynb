```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./company_sales_data - company_sales_data.csv")
x = df["month_number"]
y = df["total_profit"]
plt.plot(x,y)
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.title("Total Profit per Month")
plt.show()

ls=['-','-','-.',"--",':',"-"]
num=0
for i in df:
    if i == "month_number" or i == "total_units" or i == "total_profit" or i == "facewash":
        continue

    y = df[i]
    if i == "moisturizer":
        plt.plot(x,y, label=i+"/facewash", linestyle=ls[num])
    else:
        plt.plot(x,y, label=i, linestyle=ls[num])
    num+=1
plt.xlabel("Month")
plt.ylabel("Product Sales")
plt.title("Product Sales Data")
plt.legend()
plt.show()

column_sums = df.sum(0)
sales_sum = []
product_percent = []
labels = []
#find total units sold
for i in df:
    if i == "month_number" or i == "total_units" or i == "total_profit":
        continue
    sales_sum.append(column_sums[i])
    labels.append(i)
#percentage in pie chart
total_units = sum(sales_sum)
for i in df:
    if i == "month_number" or i == "total_units" or i == "total_profit":
        continue
    product_percent.append((column_sums[i]/total_units)*100)

fig1, ax1 = plt.subplots()
ax1.pie(product_percent, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis("equal")
plt.title("Product Sales Data for 2020")
plt.show()
```
